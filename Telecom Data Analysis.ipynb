```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
a = sns.barplot(x = 'day', y = 'total_bill',data = tips)
a.set(xlabel = 'Day_Name', ylabel = 'Total_bill', title = "The Day wise Analysis")
```

```python
a = sns.barplot(x = 'total_bill', y = 'day',data = tips)
a.set(xlabel = 'Total_bill', ylabel = 'Day', title = "The Day wise Analysis")
```

```python
a = sns.barplot(x = 'day', y = 'total_bill',hue = 'smoker',data = tips,palette= 'RdBu')
a.set(xlabel = 'Day_Name', ylabel = 'Total_bill', title = "The Day wise Analysis")
```

```python
sns.pairplot(tips,hue = 'sex',vars = ['total_bill','tip','size'])
```

```python
sns.boxplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips,palette= 'rainbow')
```
